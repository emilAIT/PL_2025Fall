```python
def bubblesort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
```

```python
def merge(arr1, arr2):
    result = [0]*(len(arr1) + len(arr2))
    i, i1, i2 = 0, 0, 0
    while i1 < len(arr1) and i2 < len(arr2):
        if arr1[0] < arr2[0]:
            result[i] = arr1[i1]
            i1 += 1
        else:
            result[i] = arr2[i2]
            i2 += 1
        i += 1
    
    while i1 < len(arr1):
        result[i] = arr1[i1]
        i, i1 = i+1, i1+1

    while i2 < len(arr2):
        result[i] = arr2[i2]
        i, i2 = i+1, i2+1

    return result

def mergesort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])
    return merge(left, right)
```

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[randint(0, len(arr)-1)]
    left, right, mid = [], [], []
    for i in arr:
        if i < pivot:
            left.append(i)
        elif i > pivot:
            right.append(i)
        else:
            mid.append(i)
    return quicksort(left) + mid + quicksort(right)
```
